```
import tensorflow as tf
import numpy as np
from tensorflow import keras
import random
```

```
def datamap(d):
  return list(map(lambda x: 1 if x == 'x' else 2 if x == 'o' else 0, d.split(',')[:-1]))
```

```
train = None
learn = None
with open("tic-tac-toe.data", 'r') as d:
  lines = d.readlines()
  random.shuffle(lines)
  train = list(map(datamap, lines))
  learn = list(map(lambda x: 1 if x.split(',')[-1:][0][:-1] == 'positive' else 0, lines))
              
print(train)
print(learn)
```

```
model = tf.keras.Sequential([
  keras.layers.Dense(units=32, input_shape=(9,), activation='relu'),
  keras.layers.Dense(units=9, activation='relu'),
  keras.layers.Dense(units=1, activation="sigmoid"),
])
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
model.fit(train, learn, epochs=500, shuffle=True)
```

```
def tttp(arr):
  print('',arr[0],arr[1],arr[2],'\n',arr[3],arr[4],arr[5],'\n',arr[6],arr[7],arr[8])
```

```
model.predict([[1,1,1,1,2,2,1,2,2], [1,1,1,1,2,2,2,1,2], [1,1,1,2,2,0,0,0,0], [1,1,1,0,0,2,0,0,2]])  # positive
```

```
model.predict([[2,2,1,1,1,2,2,1,1], [1,2,1,2,2,0,1,2,1], [0,2,1,1,2,1,1,2,2], [1,2,2,2,1,1,1,1,2]])  # negative
```

```
model.evaluate(train,  learn, verbose=2)
```
